<h2> Sorting</h2>

```python
people = {
    "first": ["sara", "tara", "zara"],
    "last": ["dar", "wani", "bhat"],
    "email": ["s@gmail.com", "t@email.com", "z@email.com"]
}
```

```python
df = pd.DataFrame(people)
df
```

```python
#Sort by a specified column:

df.sort_values(by='last')
```

```python
# Sort in descending order by that same column:

df.sort_values(by='last', ascending=False)
```

```python
# Sort by one column, then another (in descending order):

df.sort_values(by=['last', 'first'], ascending=False)
```

# *Merge*

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city")  #intersection,inner join
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")    #intersection
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")     #union
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
# indicator flag indicates elements belong to which data frame either left or right
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

# *Cocatenation*

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
# ignore index
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
#keys
df = pd.concat([india_weather, us_weather], keys=["india", "us"])
df
```

```python
df.loc["india"]
```
